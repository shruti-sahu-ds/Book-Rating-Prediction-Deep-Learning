## Phase 3: Exploratory Data Analysis (EDA)

Exploratory Data Analysis is used to understand the structure and patterns present in the dataset.

In this phase, various statistical summaries and visualizations are used to identify trends, relationships, and patterns in the data.

The analysis focuses on:

• Distribution of book prices  
• Distribution of ratings  
• Category-wise book availability  
• Relationship between price and rating  
• Correlation between numerical variables  

These insights help understand the characteristics of the book market and prepare the dataset for machine learning modeling.

```python
# Import libraries required for analysis and visualization

import pandas as pd        # Used for data manipulation and analysis
import matplotlib.pyplot as plt   # Used for creating charts
import seaborn as sns      # Used for advanced statistical visualizations
```

```python
# Load the cleaned dataset created in Phase 2

df = pd.read_csv("books_clean.csv")

# Display first few rows of the dataset
df.head()
```

```python
# Display dataset structure including column types and number of records

df.info()

# Generate statistical summary of numerical columns

df.describe()
```

```python
# Create a count plot to visualize distribution of book ratings

plt.figure(figsize=(6,4))

sns.countplot(x="Rating", data=df)

# Add title for the chart
plt.title("Distribution of Book Ratings")

# Display the plot
plt.show()
```

### Insight 3: Distribution of Book Ratings

The rating distribution shows that books are fairly evenly distributed across rating categories.

However, **1-star ratings appear slightly more frequently than other ratings**, indicating that a portion of books receive relatively low customer ratings.

Overall, most books fall between **2 and 4 stars**, suggesting moderate satisfaction levels among readers.atalog.

```python
# Create histogram to understand price distribution

plt.figure(figsize=(6,4))

sns.histplot(df["Price"], bins=20)

# Add chart labels
plt.title("Distribution of Book Prices")
plt.xlabel("Price")

# Display the chart
plt.show()
```

### Insight 4: Distribution of Book Prices

The histogram indicates that book prices are distributed between **£10 and £60**.

Most books fall within the **£20–£50 price range**, with no extreme price outliers observed in the dataset.

This suggests that the catalog mainly consists of **mid-priced books**, rather than very cheap or very expensive titles.

```python
# Count number of books in each category and visualize using bar chart

plt.figure(figsize=(10,6))

df["Category"].value_counts().plot(kind="bar")

# Add labels and title
plt.title("Number of Books by Category")
plt.xlabel("Category")
plt.ylabel("Count")

# Display chart
plt.show()
```

### Insight 5: Book Category Distribution

The category distribution chart shows that the dataset includes books from a wide range of genres.

Some categories appear significantly more frequently than others. 
For example, categories such as **Default, Nonfiction, and Sequential Art** contain the highest number of books.

This indicates that certain genres dominate the catalog, while other categories have relatively fewer titles.

```python
# Use boxplot to analyze relationship between price and rating

plt.figure(figsize=(6,4))

sns.boxplot(x="Rating", y="Price", data=df)

# Add chart title
plt.title("Price vs Rating")

# Display chart
plt.show()
```

### Insight 6: Relationship Between Price and Rating

The boxplot comparing price and rating indicates that books across different rating levels share similar price ranges.

There is **no clear trend showing that higher priced books receive better ratings**.

This suggests that book rating is likely influenced by factors other than price, such as content quality, author popularity, or genre.

```python
# Calculate correlation between numerical variables and visualize using heatmap

plt.figure(figsize=(6,4))

sns.heatmap(df[["Price","Rating","Stock_Count"]].corr(),
            annot=True,
            cmap="coolwarm")

# Add chart title
plt.title("Correlation Heatmap")

# Display chart
plt.show()
```

### Insight 7: Correlation Between Numerical Variables

The correlation heatmap shows that the relationships between numerical variables are extremely weak.

The correlation between **price and rating (0.028)** is close to zero, indicating almost no relationship.

Similarly, stock count shows negligible correlation with both price and rating.

This suggests that these variables operate largely independently within the dataset.
