## Phase 2: Data Cleaning and Preprocessing

After collecting the raw dataset through web scraping, the next step is to clean and prepare the data for analysis.

Data cleaning ensures that the dataset is consistent, accurate, and suitable for analysis and machine learning. In this phase the following tasks are performed:

- Handling missing values
- Converting categorical text values into numerical format
- Correcting data types
- Removing duplicates
- Creating new features for analysis

The cleaned dataset will then be saved for further exploratory data analysis and model building.

```python
import pandas as pd

# Load the raw scraped dataset
df = pd.read_csv("books_raw.csv")

# Preview dataset
df.head()
```

```python
# Check dataset structure
df.info()

# Summary statistics
df.describe()

# Check missing values
df.isnull().sum()
```

```python
# Convert rating text into numeric values

rating_map = {
    "One":1,
    "Two":2,
    "Three":3,
    "Four":4,
    "Five":5
}

df["Rating"] = df["Rating"].map(rating_map)
```

```python
# Convert stock count column to integer

df["Stock_Count"] = df["Stock_Count"].astype(int)
```

```python
# Convert stock count column to integer

df["Stock_Count"] = df["Stock_Count"].astype(int)
```

```python
df["Price_Band"] = pd.cut(
    df["Price"],
    bins=[0,20,40,60],
    labels=["Low","Medium","High"]
)
```

```python
df["Price_Band"] = pd.cut(
    df["Price"],
    bins=[0,20,40,60],
    labels=["Low","Medium","High"]
)
```

```python
# Save cleaned dataset

df.to_csv("books_clean.csv", index=False)

print("Clean dataset saved successfully")
```

The raw dataset collected from web scraping was processed to prepare it for analysis.

- Converted rating text values into numerical format
